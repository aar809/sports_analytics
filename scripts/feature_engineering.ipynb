```python
shot_dict = {
    1:'forehand',
    22:'backhand',
    3:'forehand chip/slice',
    24:'backhand slice',
    5:'forehand volley',
    26:'backhand volley',
    7:'smash',
    28:'backhand smash',
    9:'forehand drop shot',
    30:'backhand drop shot',
    11:'forehand lob',
    32:'backhand lob',
    13:'forehand half-volley',
    34:'backhand half-volley',
    15:'forehand swinging volley',
    36:'backhand swinging volley',
    41:'trick shot',
    99:'unknown shot',
    0:"N/A"
}

# 15. Direction (1=deuce court, 2=middle court, 3=ad court, 4=serve to wide, 5=serve to body, 6=server to T, 99=unknown)
direction_dict = {
    1:'deuce court',
    2: 'middle court',
    3:'ad court',
    4:'serve to wide',
    5:'serve to body',
    6:"server to T",
    99:"unknown"
}

shot_type_dict = {
    1:'1st serve',
    2:'2nd serve',
    3:'return', 
    4:'rally'
}

shot_outcome_dict = {
    1: "ace",
    2:"fault",
    3:"forced error",
    4:"unforced error",
    5:"winner",
    6:"service winner",
    7:"no outcome"
}

from_court_dict = {
    1:'deuce court', 
    2:'middle court',
    3:'ad court', 
    99:'unknown'
}

# to_which_court_dict = {
#     1:'deuce court',
#     2:'middle court',
#     3:'ad court',
#     99:'unknown'
# }




def shot(shot):
    shot = shot_dict[shot]
#     direction = df["Direction"]
#     direction = direction_dct[direction]
    return shot

def direction(direction):
    if direction in direction_dict.keys():
        direction = direction_dict[direction]
    else: direction = direction
    return direction

def shot_type(shot_type):
    shot_type = shot_type_dict[shot_type]
    return shot_type

def shot_outcome(shot_outcome):
    shot_outcome = shot_outcome_dict[shot_outcome]
    return shot_outcome

def shot_outcome_group(shot_outcome):
    if shot_outcome in ["ace", "service winner"]:
        shot_outcome_group = "service winner + ace"
    elif shot_outcome in ["unforced error","forced error"]:
        shot_outcome_group="error"
    else:
        shot_outcome_group = shot_outcome
    return shot_outcome_group

def from_which_court(number):
    court = from_court_dict[number]
    return court

# def name_transform(name):
#     return 

def to_which_court(num):
    to_which_court_dict = {
    1:'deuce court',
    2:'middle court',
    3:'ad court',
    99:'unknown',
    0: 'unknown'
    }
    res = to_which_court_dict[num]
    return res

#1-20 = Forehand ; 20-40 = Backhand
def shot_group(num):
    shot_group_dict = {
    1:'forehand',
    22:'backhand',
    3:'forehand',
    24:'backhand',
    5:'forehand',
    26:'backhand',
    7:'forehand',
    28:'backhand',
    9:'forehand',
    30:'backhand',
    11:'forehand',
    32:'backhand',
    13:'forehand',
    34:'backhand',
    15:'forehand',
    36:'backhand',
    41:'other',
    99:'other',
    0:"other"
    }
    res = shot_group_dict[num]
    return res

def direction_2(from_court,shot_group,to_which_court):
    if to_which_court == "middle court":
        return "down_mid"
    
    if from_court == 'deuce court':
        if shot_group == "forehand":
            if to_which_court=="deuce court":
                return "cross_court"
            if to_which_court == "ad court":
                return 'down_line'

        if shot_group == "backhand":
            if to_which_court=="deuce court":
                return 'inside_out'
            if to_which_court=="ad court":
                return 'inside_in'
    elif from_court == 'middle court':
        if shot_group == "forehand":
            if to_which_court=="deuce court":
                return "cross_court"
            if to_which_court == "ad court":
                return 'inside_out'

        if shot_group == "backhand":
            if to_which_court=="deuce court":
                return 'inside_out'
            if to_which_court=="ad court":
                return 'cross_court'
            
    elif from_court == 'ad court':
        if shot_group == "backhand":
            if to_which_court=="ad court":
                return 'cross_court'
            if to_which_court=="deuce court":
                return 'down_line'
            
        if shot_group == "forehand":
            if to_which_court=="ad court":
                return "inside_out"
            if to_which_court == "deuce court":
                return 'inside_in'
    else: 
        return "unknown"

def rally_hand_direction_outcome(shot_group,direction_2,Shot_Outcome_group):
    if shot_group == "forehand":
        if Shot_Outcome_group in ['error']:
            return "FH_Error"
        else:
            return "FH"+"_"+str(direction_2)
    elif shot_group =='backhand':
        if Shot_Outcome_group in ['error']:
            return "BH_Error"
        else:
            return "BH"+"_"+str(direction_2)
    else: 
        return '-'

    
def outcome_basic(field):
    if field in ['fault','unforced error','forced error']:
        res = 'not over'
    elif field in ['no outcome','winner','ace','service winner']:
        res='over'
    return res

# 16. To which court(1=deuce court, 2=middle court, 3=ad court, 99=unknown)
# 17. Depth(1=shallow, 2=deep, 3=very deep, 99=unknown)
# 19. Hit at what depth?(1=at net, 2=at baseline, 99=unknown)
# 21. Shot outcome(1=ace, 2=fault, 3=forced error, 4=unforced error, 5=-winner, 6=service winner, 7=no outcome)

# data["Shot_"] = data["Shot"].apply(funct)
data["Shot_"]= data["Shot"].apply(shot)
data["Direction_"] = data["Direction"].apply(direction)
data["Shot_Type_"] = data["Shot Type"].apply(shot_type)
# data["Direction_"] = data["Direction"].apply(funct)
data["Shot_Outcome_"] = data["Shot outcome"].apply(shot_outcome)
data["Shot_Outcome_group"] = data["Shot_Outcome_"].apply(shot_outcome_group)
data["From_Court"] = data["From which court"].apply(from_which_court)

data['direction_outcome']=data["Direction_"].astype(str) +"-"+ data['Shot_Outcome_group'].astype(str)


data['To which court_'] = data['To which court'].apply(to_which_court)
data['shot_group'] = data['Shot'].apply(shot_group)

data["Date"]=pd.to_datetime(data["Date"])
data['Year'] = pd.DatetimeIndex(data['Date']).year

data["match"] = str(data["Date"])+"-"+data["Player 1 name"]+"-"+data["Player 2 name"]

data['direction_2'] = data.apply(lambda x: direction_2(x['From_Court'], x['shot_group'],x['To which court_']), axis=1)

data['rally_outcome']=data.apply(lambda x: rally_hand_direction_outcome(x['shot_group'], x['direction_2'],x['Shot_Outcome_group']), axis=1)        

data['return_direction_outcome']=data["To which court_"] +"-"+ data['Shot_Outcome_']

data['outcome_basic']=data['Shot_Outcome_'].apply(outcome_basic)
data['return_direction_outcome_2']=data["To which court_"] +"-"+ data['outcome_basic']

#What are directions 7, 8, 9, 10, 0. (There's a lot of 10, and 8's, and a bit of 9's.)
data['serveDirection_outcome_basic']=data["Direction_"].astype(str) +"-"+ data['outcome_basic'].astype(str)

# columns[1] = columns[0].apply(funct)
# columns
# columns.rename({0:"Long",1:"Short"},inplace=True,axis=1)
data.head(100)

# data["Shot_"] = data["Shot"]
# data.head()
```
